```python
def check_dataframe(dataset,head = 5):
    print(10*"#" + "Shape".center(10) + 10*"#")
    print(dataset.shape)
    print("\n"+ 10*"#" + "Types".center(10) + 10*"#")
    print(dataset.dtypes)
    print("\n"+ 10*"#" + "First 5 Rows".center(20) + 10*"#")
    print(dataset.head(head))
    print("\n"+ 10*"#" + "Last 5 Rows".center(20) + 10*"#")
    print(dataset.tail(head))
    print("\n"+ 10*"#" + "Missing Values".center(20) + 10*"#")
    print(dataset.isnull().sum())
    print("\n"+ 10*"#" + "Description of Dataset".center(30) + 10*"#")
    print(dataset.describe([0,0.10, 0.25, 0.50,0.75,0.99]).T)
check_dataframe(df)
```

```python
le =LabelEncoder()
for i in range(0,5,2):
    df.iloc[:,i]= le.fit_transform(df.iloc[:,i])


df
```

```python
b = pd.DataFrame(data=df )
```

```python
X = b.drop(['Product'],axis=1)
y= b['Product']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=23)
```

```python
knn = KNeighborsClassifier(n_neighbors=9,n_jobs=-1)
knn.fit(X,y)
y_pred = knn.predict(X_test)
print(metrics.classification_report(y_test,y_pred))
```

```python
clf = RandomForestClassifier(max_depth=9, random_state=0)
clf.fit(X, y)
y_pred_rand = clf.predict(X_test)
print(metrics.classification_report(y_test,y_pred_rand))
```

```python
dtree = DecisionTreeClassifier(random_state=23)
from sklearn.model_selection import cross_val_score
cross_val_score(clf, X, y, cv=10).max()
```

```python
dtree= dtree.fit(X,y)
```

```python
y_pred_dtree  = dtree.predict(X_test)
print(metrics.classification_report(y_test,y_pred_dtree))
```

```python
from sklearn import tree
dot_data = tree.export_graphviz(dtree, out_file=None)
graph = graphviz.Source(dot_data)
```
